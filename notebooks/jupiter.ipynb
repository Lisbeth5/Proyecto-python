```python
import pandas as pd
import matplotlib.pyplot as plt

ruta_de_datos = "../data/processed/Productos_limpios.csv"
df = pd.read_csv(ruta_de_datos)

print("estadisticas basicas")
print(df.describe())

plt.hist(df['precio'], bins=20, edgecolor="black")
plt.title('Distribucion de precios')
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()

print("Productos mas caros:")
highestPrices = df.nlargest(10, 'precio')
print(highestPrices)
```
